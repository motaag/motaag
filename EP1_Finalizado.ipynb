#Fun√ß√£o para normalizar as amostras

```python
def normalizarDados(amostras):
    Xmax = max(amostras)
    Xmin = min(amostras)
    dadosNormalizados = []
    for valor in amostras:
        normalizado = (valor - Xmin) / (Xmax - Xmin)
        dadosNormalizados.append(normalizado)

    return dadosNormalizados
```

##Fun√ß√£o de Treinamento:

```python
def treinamento(tabela, d, pesoS):

    pesoInicial = pesoS
    tabela = tabela.transpose()
    n = 0.01
    epocas = 0
    tabela_len = tabela.shape[1]
    erro = True

    while erro:
        for i in range(tabela_len):
            u = np.dot(pesoS, tabela)
            y = pd.DataFrame(u)
            if not np.array_equal(y[i], d.values[i]):
                a = [tabela[i].values]
                pesoS += n * (d.values[i] - y[i].values) * a
                print(f"w: {pesoS}")
            else:
              erro = False
              print("Pesos foram ajustados!")
              return pesoS
        print(f"√âpoca: {epocas}")
        #print(f"w: {pesoS}")
        epocas += 1


    return pesoS
```

##Fun√ß√£o de Opera√ß√£o:



```python
peso = pd.DataFrame({'w0': [ 1.019146],'w1': [6.061482],'w2': [3.217473],'w3': [-7.284642]})
```

```python

def operacao(tabela,pesoS):


    tabela = tabela.transpose()
    classe_A= []
    classe_B= []
    classe_C= []
    amostraA= []
    amostraB= []

    for i in range(tabela.shape[0]):

        u = np.dot(pesoS, tabela)

        y = pd.DataFrame(u)
    for index, row in y.iterrows():
        for column in y.columns:
          valor = row[column]
          if valor <= 0: # -1
            valor = np.round(valor, 2) # Arredonda os valores para melhor visualiza√ß√£o
            classe_A.append(valor)
            amostraA.append((column))
          elif valor>=0: # 1
            valor = np.round(valor, 2)
            classe_B.append(valor)
            amostraB.append((column))


        totalA = len(classe_A)
        totalB = len(classe_B)

        resposta = pd.DataFrame({'A':[classe_A], 'B':[classe_B],'Total de A:': totalA, 'Total de B':totalB,'Amostras A':[amostraA], 'Amostras B': [amostraB]})

        return  resposta.transpose()

```
